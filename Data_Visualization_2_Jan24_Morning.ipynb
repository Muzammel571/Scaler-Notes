```python
# PLOT THE SALES TREND OF THE LONGEST RUNNING GAME!
```

```python
game_lifecycle = data.groupby("Name").agg(
    min_year = ("Year", "min"),
    max_year = ("Year", "max")
)
```

```python
game_lifecycle["Lifecycle"] = game_lifecycle["max_year"] - game_lifecycle["min_year"]
game_lifecycle.sort_values(["Lifecycle"], ascending = False)[:5]
```

```python
top3_publishers = data["Publisher"].value_counts().index[:3]
top3_genres = data["Genre"].value_counts().index[:3]
top3_platforms = data["Platform"].value_counts().index[:3]

top3_data = data.loc[
    (data["Publisher"].isin(top3_publishers)) &
    (data["Genre"].isin(top3_genres)) &
    (data["Platform"].isin(top3_platforms))
]
```

```python
top3_data.shape
```

```python
# Dodged Countplot
```

```python
sns.countplot(data = top3_data, x = "Publisher", hue = "Platform")
```

```python
# Stacked Countplot
```

```python
publisher_platform_data = pd.crosstab(index = top3_data["Publisher"], columns = top3_data["Platform"])
```

```python
publisher_platform_data.plot(kind = "bar", stacked = True)
plt.show()
```

```python
sns.boxplot(data = top3_data, y = "Global_Sales", x = "Publisher")
```

```python
sns.barplot(data = top3_data, x = "Genre", y = "Global_Sales")
```

```python
fig = plt.figure(figsize = (10, 3))

plt.subplot(2, 3, 1)
sns.scatterplot(data = top3_data, x = "NA_Sales", y = "EU_Sales")
plt.title("NA_Sales vs EU_Sales")

plt.subplot(2, 3, 3)
sns.scatterplot(data = top3_data, x = "NA_Sales", y = "JP_Sales")
plt.title("NA_Sales vs JP_Sales")

plt.subplot(2, 3, 5)
sns.scatterplot(data = top3_data, x = "NA_Sales", y = "Other_Sales")
plt.title("NA_Sales vs Other_Sales")

fig.suptitle("NA_Sales vs Others")
plt.show()
```

```python
publisher_platform_data.plot(kind = "bar", stacked = True)
plt.show()
```

```python
plt.figure(figsize = (12, 8)).suptitle("Global Sales Breakdown")

plt.subplot(2, 3, 1)
sns.scatterplot(data = top3_data, x = "Global_Sales", y = "NA_Sales")

plt.subplot(2, 3, 3)
sns.scatterplot(data = top3_data, x = "Global_Sales", y = "EU_Sales")

plt.subplot(2, 3, 4)
sns.scatterplot(data = top3_data, x = "Global_Sales", y = "JP_Sales")

plt.subplot(2, 3, 6)
sns.scatterplot(data = top3_data, x = "Global_Sales", y = "Other_Sales")

plt.subplot(1, 3, 2)
sns.barplot(data = top3_data, x = "Genre", y = "Global_Sales")

plt.show()

# Create one more plot here which spans the entire middle column
# Stacked Countplot, Global Sales > Publishers
```
