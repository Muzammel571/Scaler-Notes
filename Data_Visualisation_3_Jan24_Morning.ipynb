```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("vgsales.csv")
data.head()
```

```python
top3_publishers = data["Publisher"].value_counts().index[:3]
top3_genres = data["Genre"].value_counts().index[:3]
top3_platforms = data["Platform"].value_counts().index[:3]

top3_data = data.loc[
    (data["Publisher"].isin(top3_publishers)) &
    (data["Genre"].isin(top3_genres)) &
    (data["Platform"].isin(top3_platforms))
]
```

```python
# How can we visualise NA_Sales vs EU_Sales across multiple genres?
# NNC -> Numerical, Numerical, Categorical

# Numerical, Numerical -> Scatter Plot
# Can we add one categorical variable to scatter plot?
# YES! COLOR!

# Numerical, Categorical -> Boxplot
# Can we add one numerical variable to box plot?
# NO!
```

```python
plt.figure(figsize = (8, 6))

sns.scatterplot(data = top3_data, x = "NA_Sales", y = "EU_Sales", hue = "Genre")

plt.show()
```

```python
# How can we visualise global sales for each publisher divided into genres?
# NCC - Numerical, Categorical, Categorical

# Numerical, Categorical - Box Plot, Bar Plot
# Can we show another categorical variable on one of these plots?
# YES! COLOR!

# Categorical, Categorical - Dodged Countplot, Stacked Countplot
# We haven't delved into dodged / stacked bar plots!
```

```python
plt.figure(figsize = (10, 8))
sns.boxplot(data = top3_data, y = "Global_Sales", x = "Publisher", hue = "Genre")
plt.show()
```

```python
# How does rank affect the correlation between NA_Sales and EU_Sales?
# NNN -> Numerical, Numerical, Numerical
```

```python
plt.figure(figsize = (10, 8))

sns.scatterplot(
    data = data,
    x = "NA_Sales",
    y = "EU_Sales",
    size = "Rank",
    sizes = (1, 200)
)

plt.show()
```

```python
sns.jointplot(
    data = top3_data,
    x = "NA_Sales",
    y = "EU_Sales"
)
plt.show()
```

```python
sns.jointplot(
    data = top3_data,
    x = "NA_Sales",
    y = "EU_Sales",
    hue = "Genre"
)
plt.show()
```

```python
sns.pairplot(data = top3_data)
plt.show()
```

```python
sns.pairplot(data = top3_data, hue = "Genre")
plt.show()
```

```python
top3_data_correlation = top3_data.corr(numeric_only = True)
```

```python
top3_data_correlation
```

```python
sns.heatmap(top3_data_correlation, annot = True)
plt.show()
```

```python
sns.heatmap(top3_data_correlation, cmap = "coolwarm",  annot = True)
plt.show()
```
