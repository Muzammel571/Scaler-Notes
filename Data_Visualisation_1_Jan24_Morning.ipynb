```python
# Doing this plot in matplotlib

genre_counts = data["Genre"].value_counts()
x = genre_counts.index
y = genre_counts

plt.figure(figsize = (10, 8))
plt.title("Genre Counts")
plt.bar(x, y, width = 0.5, color = "orange")
plt.xlabel("Genres", fontsize = 12)
plt.xticks(rotation = 90)
plt.ylabel("Counts", fontsize = 12)
```

```python
# Making the same plot in seaborn

plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Genre")
plt.xticks(rotation = 45)
```

```python
# Making the same plot in seaborn

plt.figure(figsize = (10, 8))
sns.countplot(data = data, x = "Genre", order = data["Genre"].value_counts().index[::-1])
plt.xticks(rotation = 45)
```

### Pie Chart

```python
data.head()
```

```python
sales_data = data[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]]
```

```python
sales_data
```

```python
# sales_data.T.sum(axis = "columns")
```

```python
total_sales_by_region = sales_data.sum()
```

```python
total_sales_by_region
```

```python
plt.pie(
    total_sales_by_region,
    labels = total_sales_by_region.index
)
plt.show()
```

```python
plt.pie(
    total_sales_by_region,
    labels = total_sales_by_region.index,
    startangle = 90,
    explode = (0.2, 0, 0, 0),
    autopct = "%.2f%%" # Adds a % label to each area
)
plt.show()
```

```python
for i in range(5):
  plt.figure(figsize = (6, 5))
  plt.pie(
      total_sales_by_region,
      labels = total_sales_by_region.index,
      startangle = 45 * i,
      explode = (0.2, 0, 0, 0),
      autopct = "%.2f%%" # Adds a % label to each area
  )
```
